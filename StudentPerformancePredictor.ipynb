```python
from sklearn.preprocessing import MinMaxScaler
import joblib
```

```python
# Input features
X = df.drop('final_exam_score', axis=1)

# Target variable
y = df['final_exam_score']

# Verify
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)

print(X.columns)
```

```python
# Create a MinMaxScaler for the target variable
y_scaler = MinMaxScaler(feature_range=(0, 100))

# Fit and transform y
y_scaled = y_scaler.fit_transform(y.values.reshape(-1, 1))
```

```python
#Train-test-split
from sklearn.model_selection import train_test_split



X_train, X_test, y_train, y_test = train_test_split(
    X, y_scaled, test_size=0.2, random_state=42
)


print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Initialize model
model = LinearRegression()

# Train model on training data
model.fit(X_train, y_train)

```

```python
# View intercept and coefficients
model.intercept_ , model.coef_
```

```python
# Make predictions on test data
y_pred = model.predict(X_test)
```

```python
# Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# R-squared
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("R2 Score:", r2)
```

```python
# Evaluate model using MAPE
from sklearn.metrics import mean_absolute_percentage_error

y_pred_scaled = model.predict(X_test)
mape = mean_absolute_percentage_error(y_test, y_pred_scaled)
print("MAPE:", mape)
```

```python
#  Import joblib
import joblib

# Assume your trained model is named 'model' (Linear Regression in your case)
# Save the model to a file

joblib.dump(model, "final_exam_predictor_model.pkl")
joblib.dump(y_scaler, "y_scaler.pkl")


print("Model saved successfully as 'final_exam_predictor_model.pkl'")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import joblib

# Load the trained model
model = joblib.load('final_exam_predictor_model.pkl')
y_scaler = joblib.load('y_scaler.pkl')

MAX_SCORE = 350  # Maximum possible raw score

# ----------------------
# App Title
# ----------------------
st.set_page_config(page_title="Final Exam Score Predictor", page_icon="🎓", layout="wide")
st.title("🎓 Final Exam Score Predictor")
st.markdown("Predict the **final exam score** of a student based on key academic and personal factors.")

st.divider()

# ----------------------
# Input Sections
# ----------------------
st.header("👨‍🎓 Student Demographics")
col1, col2 = st.columns(2)
with col1:
    gender = st.selectbox('Gender', ['Male', 'Female'])
    age = st.number_input('Age', min_value=5, max_value=25, value=15)
    parental_education = st.selectbox('Parental Education', ['High School', 'Bachelor', 'Master', 'PhD'])
with col2:
    family_income = st.number_input('Family Income', min_value=0, value=50000)
    internet_access = st.selectbox('Internet Access', ['Yes', 'No'])
    tutor_support = st.selectbox('Tutor Support', ['Yes', 'No'])

st.divider()
st.header("📘 Academic Performance")
col3, col4 = st.columns(2)
with col3:
    previous_exam_score = st.number_input('Previous Exam Score', min_value=0, max_value=100, value=70)
    attendance_rate = st.number_input('Attendance Rate (%)', min_value=0, max_value=100, value=90)
    homework_completion_rate = st.number_input('Homework Completion Rate (%)', min_value=0, max_value=100, value=80)
with col4:
    class_participation_score = st.number_input('Class Participation Score', min_value=0, max_value=100, value=75)
    number_of_absences = st.number_input('Number of Absences', min_value=0, value=5)
    learning_hours_per_week = st.number_input('Learning Hours Per Week', min_value=0, value=10)
    extra_curricular_involvement = st.selectbox('Extra Curricular Involvement', ['No_activity', 'Low', 'Moderate', 'High'])

st.divider()

# ----------------------
# Encoding Inputs
# ----------------------
gender = 0 if gender == 'Male' else 1
internet_access = 1 if internet_access == 'Yes' else 0
tutor_support = 1 if tutor_support == 'Yes' else 0

extra_curricular_dict = {'No_activity': 0, 'Low': 1, 'Moderate': 2, 'High': 3}
extra_curricular_involvement = extra_curricular_dict[extra_curricular_involvement]

parental_education_dict = {'High School': 0, 'Bachelor': 1, 'Master': 2, 'PhD': 3}
parental_education = parental_education_dict[parental_education]

# ----------------------
# Create Input DataFrame
# ----------------------
input_data = pd.DataFrame({
    'gender': [gender],
    'age': [age],
    'parental_education': [parental_education],
    'family_income': [family_income],
    'internet_access': [internet_access],
    'previous_exam_score': [previous_exam_score],
    'attendance_rate': [attendance_rate],
    'homework_completion_rate': [homework_completion_rate],
    'class_participation_score': [class_participation_score],
    'number_of_absences': [number_of_absences],
    'extra_curricular_involvement': [extra_curricular_involvement],
    'learning_hours_per_week': [learning_hours_per_week],
    'tutor_support': [tutor_support]
})

# ----------------------
# Predict Button
# ----------------------
if st.button('Predict Final Exam Score'):
    # Predict scaled score
    scaled_pred = model.predict(input_data)
    # Convert scaled prediction back to raw score
    raw_prediction = y_scaler.inverse_transform(scaled_pred.reshape(-1, 1))[0][0]

    # Convert to percentage out of MAX_SCORE
    prediction_percent = (raw_prediction / MAX_SCORE) * 100

    # Determine grade based on percentage
    if prediction_percent >= 90:
        grade = 'A+'
    elif prediction_percent >= 80:
        grade = 'A'
    elif prediction_percent >= 70:
        grade = 'B+'
    elif prediction_percent >= 60:
        grade = 'B'
    elif prediction_percent >= 50:
        grade = 'C'
    else:
        grade = 'F'

    st.markdown("## 🏆 Prediction Result")
    st.success(f"**Predicted Final Exam Score (raw):** {raw_prediction:.2f}")
    st.success(f"**Predicted Score (scaled to 0-100%):** {prediction_percent:.2f}%")
    st.info(f"**Estimated Grade:** {grade}")

    st.write("### 📋 Input Summary")
    input_summary = input_data.copy()
    input_summary['gender'] = 'Male' if gender == 0 else 'Female'
    input_summary['internet_access'] = 'Yes' if internet_access == 1 else 'No'
    input_summary['tutor_support'] = 'Yes' if tutor_support == 1 else 'No'
    input_summary['parental_education'] = [k for k,v in parental_education_dict.items() if v == parental_education][0]
    input_summary['extra_curricular_involvement'] = [k for k,v in extra_curricular_dict.items() if v == extra_curricular_involvement][0]

    st.table(input_summary.T.rename(columns={0: "Value"}))
```
