**Student Performance Predictor for Eduquest Coaching Project**

```python
# Import pandas library for data manipulation and analysis
import pandas as pd
```

```python
#read dataset
df=pd.read_csv('/content/Student Performance Predictor for EduQuest Coaching.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df=df.dropna()
```

```python
df.info()
```

```python
#df.describe()  , df.describe().columns, df.describe().shape
```

```python
df.describe(include='all')
```

```python
#Checking distribution of values in the 'gender' column before encoding
df['gender'].value_counts()
```

```python
# Encode gender: Male as 0, Female as 1
df['gender']=df['gender'].replace({'Male':0, 'Female':1})
```

```python
# Verifying gender encoding
df['gender'].value_counts()
```

```python
#distribution of parental education levels
df['parental_education'].value_counts()
```

```python
# Encoding parental education as numbers
df['parental_education']=df['parental_education'].replace({'High School':0, 'Diploma':1, 'Graduate':2, 'Postgraduate':3})
```

```python
# Check internet access counts
df['internet_access'].value_counts()
```

```python
# Encode internet access
df['internet_access']=df['internet_access'].replace({'No':0, 'Yes':1})
```

```python
# Check extra curricular counts
df['extra_curricular_involvement'].value_counts()
```

```python
# Encode extra curricular involvement
df['extra_curricular_involvement']=df['extra_curricular_involvement'].replace({'Low':0, 'Moderate':1, 'High':2})
```

```python
# Check tutor support counts
df['tutor_support'].value_counts()
```

```python
# Encode tutor support
df['tutor_support']=df['tutor_support'].replace({'Yes':1, 'No':0})
```

```python
df.info()
```

```python
# Show all column names
df.columns
```

```python
#define x and y
#Define target variable
y=df['final_exam_score']

# Define input features
x=df[['gender', 'age', 'parental_education', 'family_income',
       'internet_access', 'previous_exam_score', 'attendance_rate',
       'homework_completion_rate', 'class_participation_score',
       'number_of_absences', 'extra_curricular_involvement',
       'learning_hours_per_week', 'tutor_support']]
```

```python
# Split data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=2529)
```

```python
# Select linear regression model
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
#Train the model
model.fit(x_train,y_train)
```

```python
# View intercept and coefficients
model.intercept_ , model.coef_
```

```python
# Make predictions on test data
y_pred = model.predict(x_test)
```

```python
# Evaluate model using MAPE
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test, y_pred)
```

```python
# The MAPE is 5.7% for this model → 5.7% average prediction error
```
